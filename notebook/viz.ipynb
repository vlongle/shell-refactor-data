```python
%matplotlib widget
import matplotlib.pyplot as plt
import matplotlib.widgets as widgets
import numpy as np
import torch

# example data
X = torch.randn(100, 1, 28, 28)
y = torch.randint(0, 9, (100,))
embed = torch.randn(100, 2)

# visualization
fig, ax = plt.subplots()
scatter = ax.scatter(embed[:, 0], embed[:, 1], c=y)
ax.set_title('2D Embeddings with Labels')
plt.colorbar(scatter)

# image display
img_fig, img_ax = plt.subplots()

# when hover over a point, show the image
def hover(event):
    if event.inaxes == ax:
        cont, ind = scatter.contains(event)
        if cont:
            # draw the image
            i = ind['ind'][0]
            # display annot using X
            img = X[i].squeeze().numpy()
            img_ax.imshow(img, cmap='gray')
            # i = ind['ind'][0]
            # ax.set_title(f'Label: {y[i]}')
            # fig.canvas.draw_idle()

fig.canvas.mpl_connect('motion_notify_event', hover)
```
